```python
def get_outliers(df, column):

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    left = Q1 - 1.5 * IQR
    right = Q3 + 1.5 * IQR

    outliers = df[(df[column] < left) | (df[column] > right)]
    
    print(f'Количество выбросов в столбце {column}: {len(outliers)}')
    print(f'Границы нормы: {max(round(left, 2), 0)} - {round(right, 2)}')

    if not outliers.empty:
        display(outliers[['job_title', 'experience_level', 'education_required', 'company_size', column]])

    return outliers
```

```python
outliers_yoe = get_outliers(df_jobs, 'years_of_experience')
```

Проверим соответствие признаков years_of_experience и experience_level.

Для начала выведем все возможные категории experience_level.

```python
df_jobs['experience_level'].unique()
```

Теперь проверим, в скольких строках опыт в годах не соответствует указанной категории.

```python
entry_mismatch = df_jobs[(df_jobs['experience_level'] == 'Entry (0-2 yrs)') & (df_jobs['years_of_experience'] > 2)]
mid_mismatch = df_jobs[(df_jobs['experience_level'] == 'Mid (3-5 yrs)') & ((df_jobs['years_of_experience'] < 3) | (df_jobs['years_of_experience'] > 5))]
senior_mismatch = df_jobs[(df_jobs['experience_level'] == 'Senior (6-9 yrs)') & ((df_jobs['years_of_experience'] < 6) | (df_jobs['years_of_experience'] > 9))]
lead_mismatch = df_jobs[(df_jobs['experience_level'] == 'Lead (10+ yrs)') & (df_jobs['years_of_experience'] < 10)]


print(f"Несоответствий в категории Entry: {len(entry_mismatch)}")
print(f"Несоответствий в категории Mid: {len(mid_mismatch)}")
print(f"Несоответствий в категории Mid: {len(senior_mismatch)}")
print(f"Несоответствий в категории Lead: {len(lead_mismatch)}")
```

Поскольку несоответствий очень много, эти строки нельзя удалять.

Попробуем разобраться, какой из признаков реально соответствует заявленной позиции. 

Для этого сгруппируем по категории опыта и для каждой найдем минимум, максимум, среднее и медиану стажа (years_of_experience).

```python
df_jobs.groupby('experience_level')['years_of_experience'].agg(['min', 'max', 'mean', 'median'])
```

Группировка показала, что значения среди категорий практически одинаковы. Это говорит о том, что столбец years_of_experience заполнен некорректно либо случайным образом.

Также вычислим корреляцию years_of_experience с annual_salary_usd. Если значения признака years_of_experience корректны, корреляция должна быть высокой, поскольку с ростом опыта, как правило, растет и заработная плата.

```python
df_jobs[['years_of_experience', 'annual_salary_usd']].corr()
```

Корреляция между years_of_experience и annual_salary_usd близка нулю. Взаимосвязь практически отсутствует, что указывает на то, что значения years_of_experience некорректны.

Рассчитаем среднюю зарплату для каждой категории опыта, чтобы убедиться, что значения в признаке experience_level корректны. Чем больше опыт, тем выше должна быть средняя зарплата.

```python
df_jobs.groupby('experience_level')['annual_salary_usd'].mean().sort_values()
```

Средняя зарплата растет с опытом. Следовательно, можно исходить из того, что значения признака experience_level корректны. Поэтому в дальнейшем в качестве опыта будет использоваться только признак experience_level.

Столбец years_of_experience останется без внимания, поскольку это ненадежный признак. Первичный анализ выявил выбросы со стажем 15 лет. Однако дальнейший анализ показал всю несогласованность столбца years_of_experience с категориями опыта. В связи с этим принято решение исключить этот признак из дальнейшей разработки сервиса.

Далее проверим соответствие зарплаты (annual_salary_usd) категориям зарплаты (salary_tier).

```python
df_jobs['salary_tier'].unique()
```

```python
entry_sal_mismatch = df_jobs[(df_jobs['salary_tier'] == 'Entry (<$100k)') & (df_jobs['annual_salary_usd'] > 100000)]
mid_sal_mismatch = df_jobs[(df_jobs['salary_tier'] == 'Mid ($100-150k)') & ((df_jobs['annual_salary_usd'] < 100000) | (df_jobs['annual_salary_usd'] > 150000))]
upper_mid_sal_mismatch = df_jobs[(df_jobs['salary_tier'] == 'Upper-Mid ($150-200k)') & ((df_jobs['annual_salary_usd'] < 150000) | (df_jobs['annual_salary_usd'] > 200000))]
senior_sal_mismatch = df_jobs[(df_jobs['salary_tier'] == 'Senior ($200-300k)') & ((df_jobs['annual_salary_usd'] < 200000) | (df_jobs['annual_salary_usd'] > 300000))]
elite_sal_mismatch = df_jobs[(df_jobs['salary_tier'] == 'Elite (>$300k)') & (df_jobs['annual_salary_usd'] < 300000)]


print(f"Несоответствий в категории Entry: {len(entry_sal_mismatch)}")
print(f"Несоответствий в категории Mid: {len(mid_sal_mismatch)}")
print(f"Несоответствий в категории Mid: {len(upper_mid_sal_mismatch)}")
print(f"Несоответствий в категории Lead: {len(senior_sal_mismatch)}")
print(f"Несоответствий в категории Lead: {len(elite_sal_mismatch)}")
```

В числовом признаке зарплаты (annual_salary_usd) и категории зарплаты (salary_tier) расхождений не наблюдается.
